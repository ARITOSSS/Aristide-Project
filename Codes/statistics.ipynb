```python
""" This module give us the win percentage of the DSSP algorithm in Minesweeper game.
    The win percentage is calculated for each difficulty level (beginner, intermediate, expert)
    by running 10,000 games for each level. Start with a random cell. """

import numpy as np
from minesweeper import MinesweeperGame
from dssp_solver import MinesweeperSolverDSSP

def main():
    """ Main function to calculate the win percentage of the DSSP algorithm in Minesweeper. """
    # Initialize the dssp solver with a Minesweeper game (9x9 with 10 mines, no GUI)
    dssp = MinesweeperSolverDSSP(MinesweeperGame(9, 9, 10, gui=False))

    # Lists to store win percentages for each difficulty level
    wins_list_beginner = []
    wins_list_intermediate = []
    wins_list_expert = []

    # Run 10,000 games for each difficulty level
    for _ in range(10):
        wins = dssp.run_games(1000, 9, 9, 10)  # Beginner level
        wins_list_beginner.append(wins)

    for _ in range(10):
        wins = dssp.run_games(1000, 16, 16, 40)  # Intermediate level
        wins_list_intermediate.append(wins)

    for _ in range(10):
        wins = dssp.run_games(1000, 16, 30, 99)  # Expert level
        wins_list_expert.append(wins)

    print("DSSP results starting with a random cell:")

    # Calculate the overall win percentages for each difficulty and print the results
    global_win_percentage = np.mean(wins_list_beginner)
    print("Beginner:")
    print(f"Win percentage for each batch: {wins_list_beginner}")
    print(f"Global win percentage of DSSP: {global_win_percentage:.2f}%\n")

    global_win_percentage_2 = np.mean(wins_list_intermediate)
    print("Intermediate:")
    print(f"Win percentage for each batch: {wins_list_intermediate}")
    print(f"Global win percentage of DSSP: {global_win_percentage_2:.2f}%\n")

    global_win_percentage_3 = np.mean(wins_list_expert)
    print("Expert:")
    print(f"Win percentage for each batch: {wins_list_expert}")
    print(f"Global win percentage of DSSP: {global_win_percentage_3:.2f}%")

if __name__ == "__main__":
    main()
```

```python
""" This script runs simulations to test the performance of the DSSP algorithm
 on Minesweeper games with different mine densities. """
import matplotlib.pyplot as plt

def main():
    """ Main function to run the simulations and plot the results. """
    # Initialize the solver with a game configuration
    dssp = MinesweeperSolverDSSP(MinesweeperGame(9, 9, 10, gui=False))
    wins_list_density = []  # List to store win percentages
    mine_densities = []  # List to store mine densities

    # Run simulations for different mine densities (up to 40%)
    for mines in range(0, 33, 5):  # Increase the number of mines by steps of 5, up to 32 mines
        win_rate = dssp.run_games(1000, 9, 9, mines)
        wins_list_density.append(win_rate)
        density = mines / (9 * 9)  # Calculate the mine density (total mines / board size)
        mine_densities.append(density)

    # Plot the results
    plt.figure(figsize=(8, 6))
    plt.plot(mine_densities, wins_list_density, marker='s',
            linestyle='-', color='g', label='DSSP', markersize=7)

    # Add labels and a title
    plt.xlabel('Mine Density', fontsize=12)
    plt.ylabel('Win Percentage', fontsize=12)
    plt.title('Win Percent vs. Mine Density with fixed Board Size (81 squares)', fontsize=14)
    plt.grid(True)
    plt.xlim(0, 0.4)
    plt.ylim(0, 100)

    # Add a legend
    plt.legend(loc='upper right')

    # Show the graph
    plt.show()

if __name__ == "__main__":
    main()
```
